```python
import os
import shutil
import re
# Source folder with your TIFF files
for i in range(1,3):
  source_dir = rf"/media/shilab/XG_Shi_SSD/XG_AE_trial2/plate6/WellC4/ff_decon_16bit/round{i}"
  # Destination parent folder
  dest_dir = source_dir
  # print(os.listdir(source_dir))
  # Loop through all files in the source directory
  for filename in os.listdir(source_dir):
    if filename.lower().endswith(".tif"):
      # Extract Tile number using regex
      match = re.search(r'decon_(.+?)_round.*?Tile_(\d+)', filename)
      # match = re.search(r"(?:_.*)?_Tile_(\d{3})", filename)
      # pattern = re.compile(r"^([A-Za-z0-9\-]+)_round", re.IGNORECASE)
      # match = re.search(pattern, filename)
      if match:
        well = match.group(1)
        tile_number = match.group(2)
        tile_folder = os.path.join(dest_dir, f"Tile_{tile_number}_{well}")
        os.makedirs(tile_folder, exist_ok=True)
        # Move the file into the tile folder
        src_path = os.path.join(source_dir, filename)
        dest_path = os.path.join(tile_folder, filename)
        shutil.move(src_path, dest_path)
  print(f"round{i} finished")
print("Done organizing TIFF files by tile number.")
```
