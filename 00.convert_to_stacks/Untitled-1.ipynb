```python
import tifffile
```

```python
tumor_488 = tifffile.imread(r"/media/shilab/XG_Shi_SSD/DAPI&488_images(60x)_Cell_boundary_training/FFPE_mouse_breast_cancer (10um)/Tumor/manual1_ch00_Tile_022_488.tif")
```

```python
#clip values above a certain intensity 
min_val = 170
max_val = 2000

tumor_488_f = tumor_488.astype("float32")
tumor_488_norm = (tumor_488_f - min_val) / (max_val - min_val)
tumor_488_norm[tumor_488_norm < 0] = 0
tumor_488_norm[tumor_488_norm > 1] = 1
```

```python
import matplotlib.pyplot as plt

plt.imsave("/media/shilab/ssd2tb/Emma_Zhang/tumor_488.png", tumor_488_norm, cmap="gray")
```
